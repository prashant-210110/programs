```python
import os
os.environ["WEATHER_API_KEY"]="efa2fd8f7a854173ee58bc2668ec5025"

from langchain_google_genai import ChatGoogleGenerativeAI

gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",   # or "gemini-1.5-pro"
    temperature=0.7,
    verbose=True,
)
```

Agents

```python
from crewai import Agent

weather_agent = Agent( 
        role="Weather Agent", 
        goal="Fetch real-time weather data for the user's requested city", 
        backstory="Specialist in retrieving accurate weather information from APIs.", 
        verbose=True, 
        allow_delegation=False, 
        llm=gemini_llm
        )
```

```python
weather_content_creator = Agent(
    role="Weather Content Creator",
    goal="Parse weather data and generate a clear, human-friendly paragraph summarizing the conditions.",
    backstory="Expert at interpreting raw weather data and turning it into easy-to-understand reports.",
    verbose=True,
    allow_delegation=False,
    llm=gemini_llm  
)
weather_content_creator
```

```python
pdf_agent = Agent(
    role="PDF Agent",
    goal="Take the generated weather summary paragraph and convert it into a downloadable PDF report.",
    backstory="Specialist in document generation, skilled at formatting text into professional PDF files.",
    verbose=True,
    allow_delegation=False,
    llm=gemini_llm
)
```

Tasks


```python
from crewai import Task

# Task 1: Fetch weather data
fetch_weather_task = Task(
    description=(
        "Take the user's input city name and call the weather API "
        "to retrieve real-time weather data including temperature, humidity, "
        "pressure, wind speed, and sky conditions."
    ),
    agent=weather_agent,
    expected_output="A structured dictionary containing the latest weather details for the requested city."
)

```

```python
weather_content_task = Task(
    description=(
        "Take the structured weather data provided by the Weather Agent "
        "and generate a clear, human-friendly paragraph summarizing the conditions. "
        "The summary should mention temperature, feels-like, humidity, pressure, wind speed, "
        "and sky conditions in a concise way."
    ),
    agent=weather_content_creator,
    expected_output="A readable weather summary paragraph suitable for end users."
)
```

```python
pdf_task = Task(
    description="Convert the generated weather summary paragraph into a well-formatted PDF report.",
    agent=pdf_agent,
    expected_output="A PDF file containing the weather summary, ready for download."
)
```

CREW

```python
from crewai import Crew
weather_crew = Crew(
    name="Weather Report Crew",
    agents=[weather_agent,weather_content_creator,pdf_agent],
    tasks=[fetch_weather_task,weather_content_task,pdf_task],
    verbose=True,
    process="sequential"
)
```

```python
result = weather_crew.kickoff({"input_city":"Hyderabad"})

print(result)
```
