```python
from langchain_openai import ChatOpenAI

openai_llm = ChatOpenAI(
    model="gpt-4o-mini",   # or "gpt-4o", "gpt-4-turbo", "gpt-3.5-turbo"
    temperature=0.7,
    verbose=True,
)

```

```python
from crewai import Agent,Crew
from phi.model.google import Gemini


translator = Agent(
    role="Language Translator",
    goal="Translate text to the user's requested language accurately",
    backstory="Expert linguist fluent in multiple languages.",
    verbose=True,
    allow_delegation=False,
    llm=openai_llm
    
)
translator
```

```python
summarizer = Agent(model=Gemini(id="gemini-2.5-flash"),    role="summaring the text",
    goal="summarize the text into main points which contain 5 to 8 bullet points,and give the output in the user required langauge",
    backstory="Skilled at distilling complex information into clear summaries.",
    verbose=True,
    allow_delegation=False,
    llm=openai_llm
)
summarizer
```

```python
from crewai import Task
translator_task = Task(
    description="Translate the following paragraph to {target_language}: {paragraph}",
    expected_output="The complete translated text in {target_language}",
    agent=translator
)
translator_task

```

```python
summarizer_task = Task(
    description="give some bullet points in {target_language} for the following paragraph: {paragraph}",
    expected_output="bullet poiont summary in {target_language}",
    agent=summarizer
)
```

```python
crew=Crew(
    agents=[translator,summarizer],
    tasks=[translator_task,summarizer_task],
    verbose=True
)
```

```python
result = crew.kickoff(inputs={
    "paragraph": "Artificial intelligence is transforming how we work and live.",
    "target_language": "Telugu"
})

print("Final Result:", result)
```
