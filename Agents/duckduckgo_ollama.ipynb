```python
import os
from phi.agent import Agent
from phi.model.ollama import Ollama

# Create the Agent
agent1 = Agent(
    name="Ollama Agent",
    model=Ollama(id="llama2"),  # or "llama3", "mistral", etc.
    instructions=[
        "Answer the user's query concisely.",
        "Provide clear explanations,in simple way"
    ],
    markdown=True,
)

# Run the Agent
response = agent1.run("WHAT is difference between agentic ai and ai agents")
print(response.content)
```

```python
response = agent1.run("prepare some python interview questions for freshers")
print(response.content)
```

```python
import os
from phi.agent import Agent
from phi.model.ollama import Ollama
from phi.tools.yfinance import YFinanceTools

# Create the Agent
agent = Agent(
    name="Stock Research Assistant",
    model=Ollama(id="llama2",stream=False),  
    instructions=[
        "Answer the user's query concisely.",
        "Provide clear explanations.",
        "just tell what the use ask",
        "Fetch live stock prices using search_finance.", 
        "Summarize market news with search_web.", 
        "Visualize trends with execute_code_orchestration_v2.", 
        "Always explain clearly and concisely."
    ],
    markdown=True,
)

# Run the Agent
response = agent.run("what is the current stock price of apple")
print(response.content)
```
