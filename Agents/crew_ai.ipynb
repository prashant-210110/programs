```python
from dotenv import load_dotenv
load_dotenv() 

import os
```

```python
from langchain_community.chat_models import ChatOllama 
ollama_llm=ChatOllama( 
                    model="llama2",
                    temperature=0.2, 
                    verbose=True )
```

```python
from crewai import Agent,Crew

translator = Agent(
    role="Language Translator",
    goal="Translate text to the user's requested language accurately",
    backstory="Expert linguist fluent in multiple languages.",
    verbose=True,
    allow_delegation=False,
    llm=ollama_llm
)
translator
```

```python
summarizer = Agent(
    role="summaring the text",
    goal="summarize the text into main points which contain 5 to 8 bullet points",
    backstory="Skilled at distilling complex information into clear summaries.",
    verbose=True,
    allow_delegation=False,
    llm=ollama_llm
)
summarizer
```

```python
from crewai import Task


translator_task = Task(
    description="Translate the following paragraph to {target_language}: {paragraph}",
    expected_output="The complete translated text in {target_language}",
    agent=translator
)
translator_task
```

```python
summarizer_task = Task(
    description="give some bullet points",
    expected_output="bullet poiont summary",
    agent=summarizer
)
summarizer_task
```

```python
from crewai import Crew


crew=Crew(
    agents=[translator,summarizer],
    tasks=[translator_task,summarizer_task],
    verbose=True
)
```

```python
result = crew.kickoff(inputs={
    "paragraph": "Artificial intelligence is transforming industries worldwide...",
    "target_language": "telugu"
})
print(result)
```
