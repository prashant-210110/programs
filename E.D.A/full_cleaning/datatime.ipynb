```python
import numpy as np
import pandas as pd
```

DatetimeFeatures            
class feature_engine.datetime.DatetimeFeatures(variables=None, features_to_extract=None, drop_original=True, missing_values='raise', dayfirst=False, yearfirst=False, utc=None, format=None)        
it extracts date and time features from datetime variables, adding new columns to the dataset. DatetimeFeatures can extract datetime information from existing datetime or object-like variables or from the dataframe index.

```python
data=pd.DataFrame(dict(
    date=["2020-09-10",np.nan,"2021-10-23","2022-02-20"]
))
data
```

```python
from feature_engine.datetime import DatetimeFeatures
model=DatetimeFeatures(features_to_extract=["year","month","quarter"],missing_values="ignore")
model.fit_transform(data)
```

DatetimeSubtraction         
class feature_engine.datetime.DatetimeSubtraction(variables=None, reference=None, new_variables_names=None, output_unit='D', missing_values='ignore', drop_original=False, dayfirst=False, yearfirst=False, utc=None, format=None)              
it applies datetime subtraction between a group of datetime variables and one or more datetime features, adding the resulting variables to the dataframe.

```python
data=pd.DataFrame(dict(
    date=["2020-09-10",np.nan,"2021-10-23","2022-02-20"],
    date2=["2021-12-10","2022-01-23","2019-04-27","2024-03-31"]
))
data
```

```python
from feature_engine.datetime import DatetimeSubtraction
modelsub=DatetimeSubtraction()
modelsub.fit_transform(data)
```
