```python
df[["churned"]].value_counts()
```

```python
#imbalance data set    doing up sampling
majority=df[df["churned"]==0]
minority=df[df["churned"]==1]
print(majority.value_counts().sum())
print(minority.value_counts().sum())
```

```python
from sklearn.utils import resample
up_sampled_miority=resample(minority,replace=True,random_state=42,n_samples=len(majority))
```

```python
up_sampled_miority.value_counts().sum()
```

```python
majority.value_counts().sum()
```

```python
#downsampling
down_sampled_majority=resample(majority,random_state=42,n_samples=len(minority))
```

```python
minority.value_counts().sum()
```

```python
down_sampled_majority.value_counts().sum()
```
