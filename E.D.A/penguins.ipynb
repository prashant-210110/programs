```python
import pandas as pd
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("penguins")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["sex"].unique()
```

```python
df["sex"]=df["sex"].fillna(df["sex"].ffill())
df["sex"].isnull().sum()
```

```python
df.columns
```

```python
df["body_mass_g"]=df["body_mass_g"].fillna(df["body_mass_g"].mean())
```

```python
df["flipper_length_mm"]=df["flipper_length_mm"].fillna(df["flipper_length_mm"].median())
```

```python
df.isnull().sum()
```

```python
df["bill_length_mm"]=df["bill_length_mm"].fillna(df["bill_length_mm"].median())
```

```python
df["bill_depth_mm"]=df["bill_depth_mm"].fillna(df["bill_depth_mm"].median())
```

```python
df.isnull().sum()
```
