```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df["town"]=encoder.fit_transform(df["town"])
```

```python
df.head()
```

```python
df.info()
```

```python
x=df[["town","area"]]
y=df["price"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=10,train_size=0.78)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(mean_absolute_error(y_test,y_predict))
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x_train,y_train)
y_predict1=model1.predict(x_test)
print(r2_score(y_test,y_predict1))
print(mean_squared_error(y_test,y_predict1))
print(mean_absolute_error(y_test,y_predict1))
```
