```python
from crewai import Agent, Task, Crew
from phi.model.ollama import Ollama
```

```python
import os
from langchain_community.chat_models import ChatOllama

ollama_llm = ChatOllama(
    model="llama3", 
    temperature=0.3
    )

```

```python
# Agent 1: Coding Agent
coding_agent = Agent(
    role="Developer",
    goal="Generate Python code based on the user's requirement.",
    backstory="Provides simple, efficient, and easy-to-understand Python code solutions.",
    model=ollama_llm,
    verbose=True,
)
```

```python
# Agent 2: Debugger
debugger = Agent(
    role="Debugger",
    goal="Fixes errors and ensures runnable code.",
    backstory="Checks the given code for bugs, syntax errors, and logic issues.",
    model=ollama_llm,
    verbose=True,
)
```

```python
# Agent 3: Reviewer
reviewer = Agent(
    role="Reviewer",
    goal="Validates code quality and correctness.",
    backstory="Reviews code for correctness, efficiency, readability, and maintainability.",
    model=ollama_llm,
    verbose=True,
)
```

```python
# Agent 4: Explainer
explaining_agent = Agent(
    role="Explainer",
    goal="Provides detailed explanations of questions.",
    backstory="Explains concepts in beginner-friendly language with examples.",
    model=ollama_llm,
    verbose=True,
)
```

```python
# Define tasks
task_input = "Write a Python function to calculate the factorial of a number."
```

```python
explain_task = Task(
    description=f"Explain the following program request in simple terms: {task_input}",
    agent=explaining_agent,
    expected_output="A beginner-friendly explanation of the requested program in plain language."
)

coding_task = Task(
    description=f"Generate Python code for: {task_input}",
    agent=coding_agent,
    expected_output="Complete Python code with clear comments, beginner-friendly and easy to run."
)

debug_task = Task(
    description="Fix any issues in the generated code.",
    agent=debugger,
    depends_on=[coding_task],
    expected_output="Corrected Python code that runs without errors and is beginner-friendly."
)

review_task = Task(
    description="Review the debugged code. Reply 'APPROVED' if valid.",
    agent=reviewer,
    depends_on=[debug_task],
    expected_output="Either 'APPROVED' with the final code block, or feedback with suggested improvements."
)
```

```python
# Crew orchestration
crew = Crew(
    agents=[coding_agent, debugger],
    tasks=[coding_task, debug_task],
    verbose=True    
)
```

```python
# Run workflow correctly
results = crew.kickoff()

print("\n--- Final Output ---\n")
print(results)
```
