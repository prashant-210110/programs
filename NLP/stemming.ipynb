Stemming in Natural Language Processing (NLP) is the process of reducing words to their root form by chopping off prefixes or suffixes

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer,RegexpStemmer

words = ["running", "runs", "runner", "connected", "connections"]
words
```

```python
porter = PorterStemmer()
for word in words:
    print(porter.stem(word))
```

```python
snowball = SnowballStemmer("english")
for word in words:
    print(snowball.stem(word))
```

```python
lancaster = LancasterStemmer()
for word in words:
    print(lancaster.stem(word))
```

```python
regex = RegexpStemmer("s$ ion$")
for word in words:
    print(regex.stem(word))
```
