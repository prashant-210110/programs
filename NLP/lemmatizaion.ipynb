Lemmatization in Natural Language Processing (NLP) is the process of reducing words to their base or dictionary form (lemma) using linguistic knowledge such as vocabulary and grammar.

```python
import nltk
nltk.download('wordnet')
```

```python
words = ["running", "runs", "runner", "connected", "connections"]
words
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
    print(lemmatizer.lemmatize(word))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
    print(lemmatizer.lemmatize(word,pos='v'))  # verb
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
    print(lemmatizer.lemmatize(word,pos='n'))  # noun
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
    print(lemmatizer.lemmatize(word,pos='a'))  # adjective
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
    print(lemmatizer.lemmatize(word,pos='r'))  # adverb
```
