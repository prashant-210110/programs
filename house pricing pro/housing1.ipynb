```python
import pandas as pd

def count_items(cell):
    if pd.isna(cell) or str(cell).strip().lower() == "none":
        return 0
    items = [x.strip() for x in str(cell).split(",")]
    return len(items)
```

```python
# Add a new column with counts
df["Amenities"] = df["Amenities"].apply(count_items)

print(df)
```

```python
x=df.drop("Price_in_Lakhs",axis=1)
y=df["Price_in_Lakhs"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#ada boost regressor model
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score
ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
y_pred_ada=ada.predict(x_test)
print(r2_score(y_test,y_pred_ada))
print(mean_absolute_error(y_test,y_pred_ada))
print(mean_squared_error(y_test,y_pred_ada))
print(root_mean_squared_error(y_test,y_pred_ada))
```
