```python
x=df.drop("price_in_lakhs",axis=1)
y=df["price_in_lakhs"]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score 
linear=LinearRegression()
linear.fit(x_train,y_train)
y_pred_lin=linear.predict(x_test)
print(r2_score(y_test,y_pred_lin))
print(mean_absolute_error(y_test,y_pred_lin))
print(mean_squared_error(y_test,y_pred_lin))
print(root_mean_squared_error(y_test,y_pred_lin))
```

```python
#support vector regressor model
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score 
svr=LinearSVR()
svr.fit(x_train,y_train)
y_pred_svr=svr.predict(x_test)
print(r2_score(y_test,y_pred_svr))
print(mean_absolute_error(y_test,y_pred_svr))
print(mean_squared_error(y_test,y_pred_svr))
print(root_mean_squared_error(y_test,y_pred_svr))
```

```python
#k nearest neighbor regressor model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score   
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
print(r2_score(y_test,y_pred_knn))
print(mean_absolute_error(y_test,y_pred_knn))
print(mean_squared_error(y_test,y_pred_knn))
print(root_mean_squared_error(y_test,y_pred_knn))
```

```python
#decision tree regressor model
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score 
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print(r2_score(y_test,y_pred_dt))
print(mean_absolute_error(y_test,y_pred_dt))
print(mean_squared_error(y_test,y_pred_dt))
print(root_mean_squared_error(y_test,y_pred_dt))
```

```python
# random forest regressor model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score 
rf=RandomForestRegressor()  
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
print(r2_score(y_test,y_pred_rf))
print(mean_absolute_error(y_test,y_pred_rf))
print(mean_squared_error(y_test,y_pred_rf))
print(root_mean_squared_error(y_test,y_pred_rf))
```

```python
#gradient boosting regressor model
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
y_pred_gb=gb.predict(x_test)
print(r2_score(y_test,y_pred_gb))
print(mean_absolute_error(y_test,y_pred_gb))
print(mean_squared_error(y_test,y_pred_gb))
print(root_mean_squared_error(y_test,y_pred_gb))
```

```python
#ada boost regressor model
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error,r2_score
ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
y_pred_ada=ada.predict(x_test)
print(r2_score(y_test,y_pred_ada))
print(mean_absolute_error(y_test,y_pred_ada))
print(mean_squared_error(y_test,y_pred_ada))
print(root_mean_squared_error(y_test,y_pred_ada))
```

```python
import pickle
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

# 1. Sample Data and Training
data = pd.DataFrame({"Score": ["Low", "Medium", "High", "Low"]})
encoder = OrdinalEncoder(categories=[["Low", "Medium", "High"]])
encoded_data = encoder.fit_transform(data)

# 2. Pickling (Saving) the fitted encoder
with open("ordinal_encoder.pkl", "wb") as f:
    pickle.dump(encoder, f)

# 3. Loading and using in production
with open("ordinal_encoder.pkl", "rb") as f:
    loaded_encoder = pickle.load(f)

new_data = pd.DataFrame({"Score": ["High", "Low"]})
new_encoded = loaded_encoder.transform(new_data)
print(new_encoded) # Outputs: [[2.], [0.]]
```
