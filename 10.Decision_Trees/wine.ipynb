```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
dataset=load_wine()
dataset
```

```python
dataset.target
```

```python
dataset.feature_names
```

```python
print(dataset.DESCR)
```

```python
cols=["Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
```

```python
dataset.data
```

```python
df=pd.DataFrame(dataset.data,columns=cols)
df["target"]=dataset.target
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
x=df.drop("target",axis=1)
y=df["target"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=40,train_size=0.8)
x_train.shape,x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
print(accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```

```python
from sklearn import tree
plt.figure(figsize=(10,6))
tree.plot_tree(model,
               max_depth=None,
               feature_names=cols,
               filled=True,
               label="all",
               impurity=True,
               rounded=True,
               fontsize=None,
               precision=1,
               node_ids=True
               )
```

post pruning

```python
model_post=DecisionTreeClassifier(max_depth=2)
model_post.fit(x_train,y_train)
tree.plot_tree(model_post,filled=True)
```

```python
parms={
    "criterion":['gini', 'entropy', 'log_loss'],
    "splitter":['best', 'random'] ,
    "max_features":['auto', 'sqrt', 'log2'],
    "max_depth":[1,2,3,4,5,6,7,8,9],
    "min_samples_split":[1,2,3,4,5,6],
    "min_samples_leaf":[1,2,3,4,5,6]
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_model=GridSearchCV(model,param_grid=parms,cv=3)
grid_model.fit(x_train,y_train)
```

```python
grid_model.get_params()
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

this is nothing but hyperparameter tuning and also called pre puning

```python
model_best=DecisionTreeClassifier(criterion="entropy",max_depth=5,max_features="log2",min_samples_leaf=3,min_samples_split=3,splitter="best")
model_best.fit(x_train,y_train)
y_predict_best=model_best.predict(x_test)
print(accuracy_score(y_test,y_predict_best))
print(classification_report(y_test,y_predict_best))
print(confusion_matrix(y_test,y_predict_best))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
tree.plot_tree(model_best,
               max_depth=None,
               feature_names=cols,
               filled=True,
               label="all",
               impurity=True,
               rounded=True,
               fontsize=6,
               precision=1,
               node_ids=True
               )
```
