```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
dataset.data
```

```python
print(dataset.DESCR)
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df["target"]=dataset.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum ()
```

```python
x=df.drop("target",axis=1)
y=df["target"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20,train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
print(r2_score(y_test,y_predict))
print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
```

```python
from sklearn import tree
tree.plot_tree(model,filled=True)
```

```python
parms={
    "criterion":['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    "splitter":['best', 'random'] ,
    "max_features":['auto', 'sqrt', 'log2'],
    "max_depth":[1,2,3,4,5,6,7,8,9],
    "min_samples_split":[1,2,3,4,5,6],
    "min_samples_leaf":[1,2,3,4,5,6]
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_model=GridSearchCV(model,param_grid=parms,cv=3)
grid_model.fit(x_train,y_train)
```

```python
model1=DecisionTreeRegressor(criterion="absolute_error",max_depth=4,max_features="log2",min_samples_leaf=6,min_samples_split=4,splitter="random")
model1.fit(x_train,y_train)
y_predict1=model1.predict(x_test)
print(r2_score(y_test,y_predict1))
print(mean_absolute_error(y_test,y_predict1))
print(mean_squared_error(y_test,y_predict1))
```

```python
tree.plot_tree(model1,filled=True)
```
