```python
#importing libraries        
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data set
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\Position_Salaries.csv")
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
df
```

```python
plt.scatter(x,y)
```

```python
#model building
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x,y)
```

```python
#visualizing the linear regression
plt.scatter(x,y,color="k")
plt.plot(x,model1.predict(x))
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_poly=PolynomialFeatures(degree=8)#degree changing there will be change in the diagram
x_poly=model_poly.fit_transform(x)
lin_2=LinearRegression()
lin_2.fit(x_poly,y)
```

```python
#under fit
plt.scatter(x,y,color="k")
plt.plot(x,lin_2.predict(model_poly.fit_transform(x)),color="red")
plt.show()
```

```python
import pandas as  pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\homeprices.csv")
df
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
plt.scatter(x,y,color="green")
plt.show()
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y,color="green")
plt.plot(x,model.predict(x),color="red")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
model2=PolynomialFeatures(degree=15)
x_poly=model2.fit_transform(x)
model3=LinearRegression()
model3.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,model3.predict(model2.fit_transform(x)))
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,model1.predict(model_poly.fit_transform(x)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,model.predict(x)))
```

```python
import numpy as np
values=np.array([[3500],[4500],[3900],[4200],[4100]])
new_predictions=model.predict(values)
new_predictions
```

```python
x_all=np.vstack((x,values))
plt.scatter(x,y,color="green")
plt.scatter(values,new_predictions,color="red")
plt.plot(x_all,model.predict(x_all))
```
