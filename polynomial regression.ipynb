```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\Position_Salaries.csv")
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x,y)
```

```python
plt.scatter(x,y,color="k")
plt.plot(x,model1.predict(x))
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=8)#degree changing there will be change in the diagram
x_poly=poly_reg.fit_transform(x)
lin_2=LinearRegression()
lin_2.fit(x_poly,y)
```

```python
#under fit
plt.scatter(x,y,color="k")
plt.plot(x,lin_2.predict(poly_reg.fit_transform(x)),color="red")
plt.show()
```
