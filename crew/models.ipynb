```python
import os
from dotenv import load_dotenv
from crewai import Agent, LLM

# Load environment variables
load_dotenv()

# Define the LLM backend (Gemini)
gemini_llm = LLM(
    model="models/gemini-1.5-flash-latest", 
    temperature=0.1)


# Create an agent using this LLM
researcher = Agent(
    role="Researcher",
    goal="Find accurate information quickly",
    backstory="A diligent and senior researcher",
    llm=gemini_llm,
    verbose=True
)

# Run the agent
query = input("Enter your query: ")
result = researcher.kickoff(query)
print(result)
```
