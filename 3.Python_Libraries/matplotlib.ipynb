```python
#scatter(x,y)
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import colors
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([50,55,65,70,75,80,85,88])
norm = colors.Normalize(vmin=0, vmax=100)
plt.scatter(x,y,color="g",edgecolors="#3F07E5",marker="*",linewidths=5,cmap="RdBu",alpha=0.9,s=[10,20,15,26,45,15,26,15],vmin=200,vmax=260)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#bar graph plot
categories = ['A', 'B', 'C']
group1 = [10, 15, 20]
group2 = [5, 10, 5]
plt.bar(categories, group1, color='skyblue', label='Group 1')
plt.bar(categories, group2, bottom=group1, color='orange', label='Group 2',zorder=5)
plt.ylabel('Total Value')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```
