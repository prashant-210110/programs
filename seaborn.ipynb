```python
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=sb.load_dataset('diamonds')
a.head()
```

```python
sb.get_dataset_names()
```

```python
a=pd.DataFrame(sb.load_dataset("tips"))
a.head()
```

```python
#we can use any one of this to the palette to get different color shades
#palette=["Set1,Set2,Set3,dark,Blues,Reds,coolwarm,viridis,deep,muted,hot,Wistiaterrain,winter,CMRmap,afmhot,binary,bone,copper,bwr,cool,summer,spring,Accent,autumn,bright,tab10,tab20,Paired,pastel,Blues_d,Oranges_d,PuOr,Purples_d,Pastel1,mako,rocket,flare,crest,icefire,vlag,Pastel2,Greens,Purples,Oranges,YlGnBu,GnBu,BuPu,,YlOrBr,GnBu,"]
#add [_r] to get reverse order,add[_d] to get dark --->this will not supports for some palettes
#one way to solve (nnormal way)
sb.barplot(x="sex",y="total_bill",data=a,hue="size",palette='Wistia_d')
plt.show()
```

```python
a.head()
```

```python
#second way-->giving custamoized colors
custom_palette = ["#FF5733", "#33FFCE", "#335BFF"]
sb.barplot(x="size",y="total_bill",data=a,palette=custom_palette)
plt.show()
```

```python
a.head()
```

```python
#third way-->sb.color_palette
c=sb.color_palette("PuOr",n_colors=4)
sb.scatterplot(x="tip",y="total_bill",data=a,palette=c,hue="day")
plt.show()
```

```python
sb.kdeplot(x="tip",y="total_bill",data=a,hue="size",palette='Wistia_d')
plt.show()
```

```python
sb.rugplot(x="sex",y="total_bill",data=a,hue="size",palette='Wistia_d')
plt.show()
```

```python
sb.jointplot(x="tip",y="total_bill",data=a,hue="size",palette='Wistia_d',kind="kde")
plt.show()
```

```python
sb.jointplot(x="sex",y="total_bill",data=a,palette='Wistia_d')
plt.show()
```

```python
sb.violinplot(x="sex",y="total_bill",data=a)
plt.show()
```

```python
sb.boxenplot(x="sex",y="total_bill",data=a)
plt.show()
```

```python
sb.ecdfplot(a["tip"],complementary=True)
plt.show
```

```python
sb.pointplot(x="sex",y="total_bill",data=a)
plt.show()
```

```python
sb.catplot(x="sex",y="total_bill",data=a,kind='strip')
plt.show()
```

```python
sb.swarmplot(x="sex",y="total_bill",data=a)
plt.show()
```
