```python
#importing libraries        
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data set
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\Position_Salaries.csv")
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
df
```

```python
plt.scatter(x,y)
```

```python
#model building
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x,y)
```

```python
#visualizing the linear regression
plt.scatter(x,y,color="k")
plt.plot(x,model1.predict(x))
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
model_poly=PolynomialFeatures(degree=8)#degree changing there will be change in the diagram
x_poly=model_poly.fit_transform(x)
lin_2=LinearRegression()
lin_2.fit(x_poly,y)
```

```python
#under fit
plt.scatter(x,y,color="k")
plt.plot(x,lin_2.predict(model_poly.fit_transform(x)),color="red")
plt.show()
```

```python
#working with Melbourne_housing_FULL data set
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\Melbourne_housing_FULL.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
cols=["Propertycount","Postcode","Distance","Bedroom2","Bathroom","Car"]
df[cols]=df[cols].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["Price"]
```

```python
#cols1=["Landsize","Building","AreaYearBuilt","Longtitude","Lattitude","CouncilArea"]
df["Landsize"]=df["Landsize"].fillna(df.Landsize.mean())
df["CouncilArea"]=df["CouncilArea"].fillna(df.Landsize.mean())
df["Longtitude"]=df["Longtitude"].fillna(df.Landsize.mean())
df["Lattitude"]=df["Lattitude"].fillna(df.Landsize.mean())
df["Price"]=df["Price"].fillna(df.Price.mean())
df["BuildingArea"]=df["BuildingArea"].fillna(df.BuildingArea.mean())
df["YearBuilt"]=df["YearBuilt"].fillna(df.Price.mean())
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce',)
x=df.drop("Price",axis=1)
y=df["Price"]
y
```
