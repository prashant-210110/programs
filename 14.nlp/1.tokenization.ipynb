Tokenization -->process of breaking text into smaller units (tokens) such as words, subwords, or sentences. it is the first step of nlp pipeline

corpus to sentences   
corpus to words             
sentences to words  

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

corpus-->paragraphs

```python
corpus="""my name is prasanth. 
I am learning data science. 
data science is an interesting field. 
it includes statistics, machine learning, deep learning etc."""
print(corpus)
corpus
#see the clear difference between corpus and print(corpus) output
```

documents-->sentences

```python
from nltk.tokenize import sent_tokenize
sentence=sent_tokenize(corpus)
sentence
```

words-->all words

```python
from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
words
```

```python
from nltk.tokenize  import RegexpTokenizer
tokenizer=RegexpTokenizer("machine learning|data science|\w+")
words=tokenizer.tokenize(corpus)
words
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
words=tokenizer.tokenize(corpus)
words
```

```python
from nltk.tokenize import BlanklineTokenizer
tokenizer=BlanklineTokenizer()
paragraphs=tokenizer.tokenize(corpus)   
paragraphs
```
