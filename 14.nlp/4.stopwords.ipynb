In Natural Language Processing (NLP), stop words are common words that usually carry little semantic meaning and are often removed during preprocessing to reduce noise in text analysis.

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize

corpus = "The cat is sitting on the mat.the cats are playing with the ball.cat name is tom."
words = word_tokenize(corpus)
words
```

```python
lword=[word.lower() for word in words]
lword
```
