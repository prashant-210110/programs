```python
#importing dataset
import pandas as pd
df = pd.read_csv(r"c:\DATA SCIENCE\Machine_Learning\Datasets\spamhamdata.csv", sep="\t", names=["label","message"],encoding="utf-8")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
#no null values
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

#text preprocessing---->stemming 
from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review=[stemmer.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
x=df["message"].values
y=df["label"].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,train_size=0.8)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#tfidf vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
Counter(y_train)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
Counter(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(x_train,y_train)
y_pred1=model1.predict(x_test)
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_train,model1.predict(x_train)))
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.svm import SVC
model2=SVC()
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
print(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_train,model2.predict(x_train)))
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(accuracy_score(y_train,model.predict(x_train)))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
y_pred3=model3.predict(x_test)
print(accuracy_score(y_test,y_pred3))
print(accuracy_score(y_train,model3.predict(x_train)))
print(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```
