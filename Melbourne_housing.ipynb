```python
#importing the libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#loading the data set
df=pd.read_csv(r"c:\DATA SCIENCE\Machine Learning\Melbourne_housing_FULL.csv")
df.head()
```

```python
df["Price"]=df["Price"].fillna(df["Price"].mean())
df["Distance"]=df["Distance"].fillna(df["Distance"].mean())
df["Bedroom2"]=df["Bedroom2"].fillna(df["Bedroom2"].mean())
df["Bathroom"]=df["Bathroom"].fillna(df["Bathroom"].mean())
df["Car"]=df["Car"].fillna(df["Car"].mean())
df["Landsize"]=df["Landsize"].fillna(df["Landsize"].mean())
df["YearBuilt"]=df["YearBuilt"].fillna(0)
df["Postcode"]=df["Postcode"].fillna(0)
df["Lattitude"]=df["Lattitude"].fillna(0)
df["Longtitude"]=df["Longtitude"].fillna(0)
df["BuildingArea"]=df["BuildingArea"].fillna(0)
df["CouncilArea"]=df["CouncilArea"].fillna(0)
df["Regionname"]=df["Regionname"].fillna(0)
df["Propertycount"]=df["Propertycount"].fillna(0)
```
