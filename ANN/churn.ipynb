```python
df.columns=df.columns.str.lower()
```

```python
num_cols=df.select_dtypes(include=["int","float"])
cat_cols=df.select_dtypes(include=["object"])
```

```python
df.columns
```

df.drop(['rownumber', 'customerid', 'surname'],inplace=True,axis=1)

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat_cols:
    df[i]=le.fit_transform(df[i])
```

```python
cat_cols
df.head()
df.dtypes
```

```python
x=df.drop("exited",axis=1)
y=df["exited"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,train_size=0.8)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
x_test_scaled
```

ANN model classification

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
model=Sequential()
#adding layers
model.add(Dense(32, activation='relu', input_dim=x_train_scaled.shape[1]))#input layer with relu activaion function with 32 neurons
model.add(Dense(16,activation='relu'))                                    #hidden layer with relu activation function with 16 neurons
model.add(Dense(1,activation='sigmoid'))                                  #output layer with sigmoid activation function with 1 neuron
```

```python
x_train_scaled.shape[1]
```

```python
model.summary()
#13*32+32
#32*16+16
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(x_train_scaled, y_train, epochs=50, batch_size=20, validation_split=0.2, verbose=1)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
y_pred=model.predict(x_test)
print(f"y predicted values are {y_pred[:10]}")
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
