```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_biclusters

# Generate biclustered data
data, rows, cols = make_biclusters(
    shape=(10, 20), 
    n_clusters=3, 
    noise=0.1, 
    random_state=40
)

# Plot the heatmap
plt.figure(figsize=(8, 6))
plt.matshow(data, cmap="viridis", fignum=1)
plt.colorbar()

# Overlay bicluster regions
for i in range(rows.shape[0]):
    row_idx = np.where(rows[i])[0]
    col_idx = np.where(cols[i])[0]
    for r in row_idx:
        for c in col_idx:
            plt.gca().add_patch(
                plt.Rectangle((c-0.5, r-0.5), 1, 1, 
                              fill=False, edgecolor="red", linewidth=1)
            )

plt.title("Biclustered Data Heatmap")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()

```
