```python
import matplotlib.pyplot as plt
```

```python
#plot a simple line chart using matplotlib for manths and sales data with customizations

months=["jan","feb","mar","apr","may","jun"]
sales=[100,600,300,100,500,150]
plt.plot(months,sales, 
         marker="o",
         color="k",
         linestyle="--",
         linewidth=2,
         markersize=10,
         markerfacecolor="r",
         markeredgecolor="b",
         alpha=0.5,
         label="sales")
plt.xlabel("months")
plt.ylabel("sales")
plt.title("sales per month")
plt.legend()
plt.grid()
plt.show()
```

```python
#create a bar chart using matplotlib to compare revenue of different products
products=["A","B","C","D","E"]
sales=[100,600,300,100,500]
plt.bar(products,sales)
plt.show()
```

```python
#geneate a histogram using matplotlib to visualize the distribution of ages 
ages=[20,30,40,50,60,70,80,90]
plt.hist(ages, bins=5)
plt.show()
```

```python
#create a scatter plot in matplotlib to visualize the relationship between advertising and sales
ad=[20,30,40,50,60,70,80,90]
sales=[100,600,300,100,500,150,200,400]
plt.scatter(ad,sales)
plt.show()
```

```python
#plot multiple lines in one matplotlib graph to compare sales of two products over the years
year=[2010,2011,2012,2013,2014]
sales1=[100,600,300,100,500]
sales2=[150,200,400,250,350]
plt.plot(year,sales1, label="sales1")
plt.plot(year,sales2, label="sales2")
#different customizations
plt.xlabel("year")
plt.ylabel("sales")
plt.title("sales comparison")
plt.legend()
plt.show()
#saving the figure
plt.savefig("sales.png")
```

```python
#create a matplotlib boxplot to analyze salary disribution in a company
salary=[30000,40000,50000,60000,70000,120000]
plt.boxplot(salary)
```

```python
cat=["q1","q2","q3","q4"]
a=[10,20,30,40]
b=[15,25,35,45]
plt.bar(cat,a, label="a")
plt.bar(cat,b,bottom=a, label="b")
plt.legend()
plt.show()
plt.style.use('ggplot')
```
