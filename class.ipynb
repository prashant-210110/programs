```python
import subprocess  #used to run like in terminal without using the terminal ....in terminal we need to type ollama list
def list_ollama():
    result=subprocess.run(['ollama', 'list'], capture_output=True, text=True)
    print(result.stdout)
list_ollama()
    
```

```python
def ask_llama(prompt:str):
    result=subprocess.run(
        ['ollama', 'run', 'tinyllama'],
        input=prompt,
        capture_output=True, 
        text=True)
    return result.stdout.strip()
ask_llama(prompt="What is the capital of India?")
```
