mean median imputation  
-->replacing the missing value with mean or mode based on the given data    
-->mean-for normal distribution data(having good bell curve)    
-->median-for not good distributed or if the data is having the outliers (not having good bell curve)

```python
df=pd.DataFrame(dict(
    x=[15,52,np.nan,53,21],
    y=["a","b","c","d","e"]
))
```

```python
df.dtypes
```

MeanMedianImputer(imputation_method='median' or 'mean', variables=None)     
-->fit(x,y=None)                               -Learn the mean or median values     
-->fit_transform(x,y=None,**fit_params)        -Fit to data, then transform it      
-->get_feature_names_out(input_features=None)  -Get output feature names for transformation     
-->get_metadata_routing()                      -Get metadata routing of this object     
-->get_params(deep=True)                       -Get parameters for this estimator       
-->set_params(**params)                        -Set the parameters of this estimator        
-->transform(X)                                -Replace missing data with the learned parameters        

```python
data=MeanMedianImputer()
data.fit_transform(df)
```

```python
data.get_metadata_routing()
```

```python
data.fit(df)
data
```

```python
data.transform(df)
```

```python
data.get_params()
```

```python
data.set_params()
```

```python
data.get_feature_names_out()
```

feature_engine.imputation.CategoricalImputer(imputation_method='missing', fill_value='Missing', variables=None, return_object=False, ignore_format=False)       
it replaces missing data in categorical variables by an arbitrary value or by the most frequent category        


```python
data_cat=pd.DataFrame(dict(
    x=[1,2,3,np.nan,5,6],
    y=["a","b",np.nan,"a",np.nan,"b"]))
```

```python
from feature_engine.imputation import CategoricalImputer
model_cat=CategoricalImputer()
model_cat.fit_transform(data_cat)
```

addingmissing imputation

it adds binary variables that indicate if data is missing.The AddMissingIndicator() works for both numerical and categorical variables.         
-->fit(x,y=None)                               -Learn the mean or median values     
-->fit_transform(x,y=None,**fit_params)        -Fit to data, then transform it      
-->get_feature_names_out(input_features=None)  -Get output feature names for transformation     
-->get_metadata_routing()                      -Get metadata routing of this object     
-->get_params(deep=True)                       -Get parameters for this estimator       
-->set_params(**params)                        -Set the parameters of this estimator        
-->transform(X)                                -Replace missing data with the learned parameters

```python
data1= pd.DataFrame(dict(
       x1 = [np.nan,1,1,0,np.nan],
       x2 = ["a", np.nan, "b", np.nan, "a"],
       ))
data1
```

```python
from feature_engine.imputation import AddMissingIndicator
model_add=AddMissingIndicator()
model_add.fit_transform(data1)
```

DropMissingData             
class feature_engine.imputation.DropMissingData(missing_only=True, threshold=None, variables=None)  
DropMissingData() deletes rows containing missing values.

```python
from feature_engine.imputation import DropMissingData
model_drop=DropMissingData()
model_drop.fit_transform(data1)
```
