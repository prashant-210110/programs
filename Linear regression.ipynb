```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#creating the data sets
study_time=np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
test_scores=np.array([50,55,65,70,75,80,85,88])
```

```python
#splitting the data
x_train,x_test,y_train,y_test=train_test_split(study_time,test_scores,test_size=0.2,random_state=0)
```

```python
#train the linear regression model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#visualization the regression line for training data
plt.scatter(x_train,y_train,color="red",label="training data")
plt.plot(x_train,model.predict(x_train),color="blue")
plt.grid()
plt.legend()
plt.show()
```

```python
#make prediction on testing 
y_predict=model.predict(x_test)
print(x_test)
print(y_predict)
```

```python
#visualizing predictions for testing data
plt.scatter(x_test,y_test,color="k",label="regression line")
plt.plot(x_train,model.predict(x_train))
plt.grid()
plt.show()
```

```python
#evaluating the model perfromance
mse=mean_squared_error(y_test,y_predict)
r2=r2_score(y_test,y_predict)
print(f"{mse:.2f}")
print(f"{r2:.2f}")
```
