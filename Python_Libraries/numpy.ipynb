```python
import numpy as np
a=np.arange(0,22,3).reshape(4,2)
print(a)
```

```python
b=np.array([[[1,5,2,3,6]]])
print(b)
print(b.ndim)
```

```python
a=np.array([10,20,30,40,50,60,70])
print(a[-1:-5:-1])
print(np.sqrt(a))
print(np.mean(a))
print(np.median(a))
print(np.var(a))
print(np.std(a))
print(np.sum(a))
print(max(a))
print(min(a))
print(np.cumsum(a))
print(np.cumprod(a))
```

```python
a=np.array([10,20,30,40,50,60,70,80])
b=np.array([11,12,13,14,15,16,17,18])
print(np.concatenate((a,b)))
print(np.stack((a,b),axis=0))
print(np.stack((a,b),axis=1))
print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
b=a.reshape(2,4)
print(b)
print(np.transpose(b))
```

```python
b=a.reshape(2,4)
print(b)
print(np.split(b,2))
print(np.hsplit(b,2))
print(np.vsplit(b,2))
```

```python
c=np.array([1,2,3,4])
print(np.repeat(c,2))
print(np.tile(c,2))
```

```python
def square (x):
    return x*x
v=np.vectorize(square)
a=np.array([1,2,3,4,5])
print(v(a))
```
