```python
import pickle
with open("model_cnn.pkl", "rb") as f:
    model= pickle.load(f)
```

```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

# Load image and convert to grayscale
img = Image.open(r"c:\Users\prasa\OneDrive\Documents\Desktop\Desktop\mnist testing pics\Screenshot 2026-02-16 212647.jpg").convert("L")

# Invert colors (MNIST digits are white on black)
img = ImageOps.invert(img)
plt.imshow(img, cmap='gray') 
plt.axis('off') 
plt.show()
```

```python
# Resize to 28x28
img = img.resize((28, 28))

# Convert to array and normalize
img_array = np.array(img) / 255.0

# Reshape for CNN input
img_array = img_array.reshape(1, 28, 28, 1)
```

```python
plt.imshow(img, cmap='gray') 
plt.axis('off') 
plt.show()
```

```python
prediction = model.predict(img_array)
predicted_class = np.argmax(prediction)
print("Predicted digit:", predicted_class)
```
